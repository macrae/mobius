## Curtis Institute of Music
#### Data Quality Assurance

Curtis provided two csv's for a $25K/1-year major gift PTG model.

Unlike the standard `constituents` and `gifts` files typically received for PTG models, for some reason unbeknownst to me, the two files are both based on the one-to-many join between `constituents.constituent_id` and `gifts.constituent_id` (hence `constituent_id` is not unique in either of these files) - and then they partitioned the table along the column-index - with one file containing some fields and the other containing other fields. Most likely there's a file output limit of some sort they are side-stepping.

### Analyze organization donor rate. 

Gifts from organizations are identified by the `Gf_CnBio_Key_Indicator` field, where the value is `O`. 

```python
n_constituents = len(constituents["Gf_CnBio_Import_ID"].unique())
n_constituents_orgs = len(constituents[constituents["Gf_CnBio_Key_Indicator"] == "O"]["Gf_CnBio_Import_ID"].unique())
n_constituents_org_rate = 100 * (n_constituents_orgs / n_constituents)

f"The number of unique constituents is {n_constituents:,} of which {n_constituents_orgs:,} are organizations, or {n_constituents_org_rate:.2f}%"
```

We want to make sure to remove these records since we are training a model to find individual major donors using Windfall household features. Including an organization doesn't make sense within the definition of the ML-task and feature space.

```python
# remove organizations
constituents, gifts = constituents[constituents["Gf_CnBio_Key_Indicator"] != "O"], gifts[gifts["Gf_CnBio_Key_Indicator"] != "O"]
```

```python
n_constituents = len(gifts["Gf_CnBio_Import_ID"].unique())

f"The number of unique individual constituents is {n_constituents:,}"
```

### Analyze Donor Experience

#### Frequency

```python
# cast `Gf_Amount` from string to float
gifts["Gf_Amount"] = round(gifts["Gf_Amount"].str.replace("$", "").str.replace(",", "").astype(float), 2)
```

```python
n_donation_mean = gifts.groupby("Gf_CnBio_Import_ID")["Gf_Amount"].apply(len).mean()
n_donation_median = gifts.groupby("Gf_CnBio_Import_ID")["Gf_Amount"].apply(len).median()

f"The average number of donations a constituent makes is {n_donation_mean:,.1f}; the median is {n_donation_median}"
```

```python
gifts.groupby("Gf_CnBio_Import_ID")["Gf_Amount"].apply(len).plot(kind="hist", bins=80, figsize=(12, 8), title="Frequency of Donations by Donor");
```

Note the density of 1-time donors. From a modeling perspective, we should experiment with regularizing donor experience by enriching the data. Biasing too heavily on donor experience generally negatively impacts the models ability to identify 1st time donors from a pool of non-donors.

To further hone our intuition, let's filter on constituents having greater than $25K

```python
ideal_ids = gifts.groupby("Gf_CnBio_Import_ID")["Gf_Amount"].sum()[
    gifts.groupby("Gf_CnBio_Import_ID")["Gf_Amount"].sum() > 25_000].index
```

```python
gifts[gifts["Gf_CnBio_Import_ID"].\
        apply(lambda x: x in ideal_ids)].\
        groupby(["Gf_CnBio_Import_ID"]).\
        apply(len).\
        plot(kind="hist", bins=40, figsize=(12, 8), title="Frequency of Donations by Ideal Donor");
```

For constituents having greater than $25K in total lifetime giving, there's still a density of 1-time donors. That means most of the people we are looking for have no donation history up until the point in time that they write a check for a major gift. I am worried about biasing too heavily on donor experience here for sure!

```python
n_ideal = len(ideal_ids)
n_constituents = len(gifts["Gf_CnBio_Import_ID"].unique())
hit_rate = 100 * (n_ideal / n_constituents)

f"There are {n_ideal} of {n_constituents:,} constituents who have donated $25K or more in their lifetime, or {hit_rate:.3}%"
```

The classes are extremely imbalanced - which makes for another opportunity for enriching the ideal populatiion with other households from accounts having similar donor experience.

```python
sql = """
DECLARE target_period_start DATE;

CREATE TEMP FUNCTION
      get_target_period_start() AS ((
        SELECT
          MAX(CAST(PARSE_TIMESTAMP("%m/%d/%Y", gf_date) AS DATE))
        FROM
           tranquil-garage-139216.portal_model_scoring.training__account_641_20210728 cand, cand.donations g ));

CREATE TEMP FUNCTION
    in_period(m1 INT64, m2 INT64, trans_date DATE, col ANY TYPE) AS (
            CASE WHEN trans_date < DATE_SUB(get_target_period_start(), INTERVAL m1 month)
                    AND trans_date >= DATE_SUB(get_target_period_start(), INTERVAL m2 month)
            THEN col
            END);

CREATE TEMP FUNCTION
    before_period(period_months INT64, trans_date DATE, col_to_sum ANY TYPE) AS (
            CASE WHEN trans_date < DATE_SUB(get_target_period_start(), INTERVAL period_months month)
                    THEN col_to_sum
                    END);

-- SET target_period_start = get_target_period_start();


WITH 
    all_constituents AS (
        SELECT
            distinct(cand.candidate_id),
            row_number() OVER (PARTITION BY candidate_id) AS rownumber,
            constituents.*
        FROM tranquil-garage-139216.portal_model_scoring.training__account_641_20210728 cand
        JOIN cand.constituents as constituents
        WHERE gf_cnbio_key_indicator != "O"
        ),

    dedup_constituents AS (  -- 50,010 distinct candidate ids matches w/ expectations
        SELECT * EXCEPT (rownumber) from all_constituents where rownumber=1
        ),
        
    constituents AS(  -- there is approximately a 47.5% match rate
        SELECT
            distinct(cand.candidate_id),
            m.windfall_id,
            cand.*EXCEPT(candidate_id)
        FROM dedup_constituents cand
        LEFT JOIN tranquil-garage-139216.portal.match as m ON m.candidate_id = cand.candidate_id
        AND m.account_id = 641
        AND m.confidence >= 0.5
    ),

    all_gifts AS (  -- 50,010 distinct candidate ids matches w/ expectations
        SELECT
            distinct(cand.candidate_id),
            row_number() OVER (PARTITION BY candidate_id, gf_date, gf_amount) AS rownumber,
            CAST(REGEXP_REPLACE(gf_amount, r'\$|,', '') as FLOAT64) AS gf_amount,
            donations.* EXCEPT(gf_amount)
        FROM tranquil-garage-139216.portal_model_scoring.training__account_641_20210728 cand
        LEFT JOIN cand.donations as donations ON candidate_id = donations.gf_cnbio_import_id
        WHERE gf_cnbio_key_indicator != "O"
        ),
        
    dedup_gifts AS (  -- remove donations per candidate that have the same value & date
        SELECT * EXCEPT (rownumber) from all_gifts where rownumber=1
        ),

    base AS (  -- 11,846 unique donors returned after filtering on gift type
        SELECT
            cand.candidate_id,
            constituents.windfall_id,
            CONCAT("cand_", cand.candidate_id) AS split_col, -- keep things household level as able
            constituents.alumnus_a_ AS alumnus,
            constituents.trustee_ AS trustee,
            gifts.gf_cnpral_school_name AS school_name,
            gifts.gf_cnpral_date_entered AS date_entered,
            gifts.gf_cnpral_date_graduated AS date_graduated,
            gifts.gf_cnpral_class_of AS class_of,
            gifts.gf_cnpral_campus AS artistic_director,
            gifts.gf_cnpralmaj_1_01_tableentriesid AS major,
            gifts.gf_amount AS gift_amount,
            gifts.gf_type AS gift_type,
            gifts.gf_date AS gift_date,
            gifts.* EXCEPT(gf_type,gf_date, gf_amount, candidate_id)
        from tranquil-garage-139216.portal_model_scoring.training__account_641_20210728 cand
        LEFT JOIN dedup_gifts as gifts ON gifts.candidate_id = cand.candidate_id
        LEFT JOIN constituents ON constituents.candidate_id = cand.candidate_id
        WHERE gifts.gf_type IN ("Cash", "Pay-Cash", "Stock/Property", "Other", "Pay-Stock/Property", "Stock/Property (Sold)", "Pay-Stock/Property (Sold)", "Recurring Gift Pay-Cash")
        ),
        
    enriched_base as (  -- 11,846 (i don't understand why we don't just do this in the CTE above)
        SELECT
            CAST(PARSE_TIMESTAMP("%m/%d/%y", base.gift_date) AS DATE) as gift_date,
            CASE
                WHEN base.gift_amount IS NULL THEN 0
                ELSE CAST(base.gift_amount AS FLOAT64)
                END AS gift_amount,
            base.* EXCEPT(gift_date, gift_amount)
        from base
        ),

    rolling_windows AS (  -- 11,846 unique constituents after filtering on gift date greater than 01/01/2000
        SELECT
            d.split_col,
            d.candidate_id,
            d.windfall_id,
            d.gift_amount,
            d.gift_date,
            SUM(d.gift_amount) OVER(PARTITION BY d.split_col
                        ORDER BY UNIX_SECONDS(TIMESTAMP(d.gift_date))
                        RANGE BETWEEN 31536000 PRECEDING AND CURRENT ROW) AS rolling_1year_amount,
            COUNT(d.split_col) OVER(PARTITION BY d.split_col
                ORDER BY UNIX_SECONDS(TIMESTAMP(d.gift_date))
                RANGE BETWEEN 31536000 PRECEDING AND CURRENT ROW) AS rolling_1year_count,
        FROM enriched_base d
--         WHERE d.gift_date > "2000-01-01"
        ),

    ideals AS( -- 242 ideals
        SELECT
            r.split_col AS candidate_id,
            ANY_VALUE(r.windfall_id) as windfall_id,
            MAX(r.rolling_1year_amount) as max_1year_gift,
            MIN(r.rolling_1year_amount) as min_1year_gift,
            "ideal" as label,
            DATE_SUB(MIN(r.gift_date), INTERVAL 1 YEAR) AS period_start,
            MIN(r.gift_date) AS period_end -- first gift that put the candidate into the ideal bucket?
        FROM rolling_windows as r
        WHERE r.rolling_1year_amount >= 25000 --THIS ^^
        # AND windfall_id IS NOT NULL
        GROUP BY r.split_col
        ),

    non_ideals AS( -- 11,604 non-ideals
        SELECT
            DISTINCT r.split_col as candidate_id ,
            ANY_VALUE(r.windfall_id) as windfall_id,
            MAX(r.rolling_1year_amount) as max_1year_gift,
            MIN(r.rolling_1year_amount) as min_1year_gift,
            "non_ideal" AS LABEL,
            DATE_SUB(target_period_start, INTERVAL 1 YEAR) AS period_start,
            target_period_start AS period_end
        FROM rolling_windows as r
        WHERE r.split_col  NOT IN (SELECT candidate_id FROM ideals) --speed up w/ join
        GROUP BY r.split_col
        ),

    union_targets AS (  -- 11,846 unique customer ids - yay!
        SELECT * FROM
        non_ideals
        UNION ALL
        SELECT * FROM
        ideals
        ),

    customer_features AS(  -- 11,846 unique customer ids
        SELECT DISTINCT 
        union_targets.candidate_id, 
        union_targets.windfall_id, 
        base.alumnus,
        base.trustee,
        CASE WHEN base.school_name = "Curtis Institute of Music" THEN 1 ELSE 0 END AS curtis_student,
        EXTRACT(YEAR FROM SAFE.PARSE_DATE("%m/%d/%Y", base.date_entered)) AS year_entered,
        EXTRACT(YEAR FROM SAFE.PARSE_DATE("%m/%d/%Y", base.date_graduated)) AS year_graduated,
        base.class_of,
        base.artistic_director,
        base.major,
        period_start, 
        period_end, 
        label,
        MAX(base.gift_amount) OVER (PARTITION BY union_targets.candidate_id, union_targets.windfall_id, period_start, period_end, label) AS max_gift_amount,
        SUM(CASE
                WHEN gift_date >= period_start AND gift_date <= period_end THEN 1 END
                ) OVER (PARTITION BY union_targets.candidate_id, union_targets.windfall_id, period_start, period_end, label) AS target_period_ct,
        SUM(CASE
                WHEN gift_date >= period_start AND gift_date <= period_end THEN gift_amount END
                ) OVER (PARTITION BY union_targets.candidate_id, union_targets.windfall_id, period_start, period_end, label) AS target_period_amt,
        SUM(CASE
                WHEN gift_date < DATE_SUB(period_start, INTERVAL 0 month)
                    AND gift_date >= DATE_SUB(period_start, INTERVAL 12 month)
                        THEN gift_amount END
            ) OVER (PARTITION BY union_targets.candidate_id, union_targets.windfall_id, period_start, period_end, label) AS one_yr_prior_amt,
        SUM(CASE
                WHEN gift_date < DATE_SUB(period_start, INTERVAL 0 month)
                    AND gift_date >= DATE_SUB(period_start, INTERVAL 36 month)
                        THEN gift_amount END
            ) OVER (PARTITION BY union_targets.candidate_id, union_targets.windfall_id, period_start, period_end, label) AS three_yr_prior_amt,
        SUM(CASE
                WHEN gift_date < DATE_SUB(period_start, INTERVAL 0 month)
                    AND gift_date >= DATE_SUB(period_start, INTERVAL 12 month)
                        THEN 1 END
            ) OVER (PARTITION BY union_targets.candidate_id, union_targets.windfall_id, period_start, period_end, label) AS one_yr_prior_ct,
        SUM(CASE
                WHEN gift_date < DATE_SUB(period_start, INTERVAL 0 month)
                    AND gift_date >= DATE_SUB(period_start, INTERVAL 36 month)
                        THEN 1 END
            ) OVER (PARTITION BY union_targets.candidate_id, union_targets.windfall_id, period_start, period_end, label)  AS three_yr_prior_ct,
        SUM(CASE
                WHEN gift_date < DATE_SUB(period_start, INTERVAL 0 month)
                    AND gift_date >= DATE_SUB(period_start, INTERVAL 60 month)
                        THEN 1 END
            ) OVER (PARTITION BY union_targets.candidate_id, union_targets.windfall_id, period_start, period_end, label)  AS five_yr_prior_ct_rfm,
        SUM(CASE
                WHEN gift_date < DATE_SUB(period_start, INTERVAL 0 month)
                    AND gift_date >= DATE_SUB(period_start, INTERVAL 60 month)
                        THEN gift_amount END
            ) OVER (PARTITION BY union_targets.candidate_id, union_targets.windfall_id, period_start, period_end, label)  AS five_yr_prior_amt_rfm,
        MIN(CASE
                WHEN gift_date < period_start
                    AND gift_date >= "2000-01-01"
                        THEN DATE_DIFF(period_start, gift_date, month) END
            ) OVER (PARTITION BY union_targets.candidate_id, union_targets.windfall_id, period_start, period_end, label)  AS recent_donation_months_before_target,
        FROM union_targets
        JOIN enriched_base base ON base.split_col = union_targets.candidate_id
        ),

    wf AS ( -- 16,737  unique candidate ids (because there was a previous Curtis data drop last year there are more unique ids than included in just the most recent)
        SELECT DISTINCT wf.*
        FROM tranquil-garage-139216.curtis.wf_features_641_20210728 wf
        JOIN tranquil-garage-139216.portal.match m ON wf.id = m.windfall_id
        WHERE m.confidence >= 0.5 AND account_id = 641
        )

SELECT  -- 11,846 unique candidate ids returned
    CASE WHEN windfall_id IS NULL THEN candidate_id ELSE windfall_id END AS windfall_id,
    ANY_VALUE(candidate_id) AS candidate_id,
    COUNT(candidate_id) AS candidate_ids,
    SUM(CASE WHEN trustee = "Trustee" THEN 1 ELSE 0 END) AS trustee,
    CASE WHEN SUM(CASE WHEN label = "ideal" THEN 1 ELSE 0 END) > 0 THEN "ideal" ELSE "non_ideal" END AS label,
    SUM(max_gift_amount) AS max_gift_amount,
    SUM(target_period_ct) AS target_period_ct,
    SUM(target_period_amt) AS target_period_amt,
    SUM(one_yr_prior_amt) AS one_yr_prior_amt,
    SUM(three_yr_prior_amt) AS three_yr_prior_amt,
    SUM(one_yr_prior_ct) AS one_yr_prior_ct,
    SUM(three_yr_prior_ct) AS three_yr_prior_ct,
    SUM(five_yr_prior_ct_rfm) AS five_yr_prior_ct_rfm,
    SUM(five_yr_prior_amt_rfm) AS five_yr_prior_amt_rfm,	
    SUM(recent_donation_months_before_target) AS recent_donation_months_before_target,
    MAX(logNetWorth) AS logNetWorth, 
    MAX(logAccreditedNetWorth) AS logAccreditedNetWorth, 
    MAX(recentPurchaseDate) AS recentPurchaseDate,
    MAX(recentDivorceDate) AS recentDivorceDate, 
    MAX(recentDeathDate) AS recentDeathDate, 
    MAX(isBoatOwner) AS isBoatOwner, 
    MAX(isPlaneOwner) AS isPlaneOwner, 
    MAX(recentPhilanthropicGiftDate) AS recentPhilanthropicGiftDate, 
    MAX(recentFoundationAssociationDate) AS recentFoundationAssociationDate, 
    MAX(isFoundationOfficer) AS isFoundationOfficer, 
    MAX(recentFoundationTrusteeDate) AS recentFoundationTrusteeDate, 
    MAX(hasFoundationAssociation) AS hasFoundationAssociation, 
    MAX(hasCharityBoardMember) AS hasCharityBoardMember, 
    MAX(hasCharityOfficer) AS hasCharityOfficer, 
    MAX(isMultiPropertyOwner) AS isMultiPropertyOwner, 
    MAX(isRentalPropertyOwner) AS isRentalPropertyOwner,
    MAX(isHouseholdDebt) AS isHouseholdDebt, 
    MAX(logTotalHouseholdDebt) AS logTotalHouseholdDebt, 
    MAX(recentSecTransactionAcquired) AS recentSecTransactionAcquired, 
    MAX(recentSecTransactionDisposed) AS recentSecTransactionDisposed, 
    MAX(recentPoliticalContributionDate) AS recentPoliticalContributionDate, 
    MAX(isSmallBusinessOwner) AS isSmallBusinessOwner, 
    MAX(isCharityBoardMember) AS isCharityBoardMember, 
    MAX(isCharityOfficer) AS isCharityOfficer, 
    MAX(primaryPropertyLoanToValue) AS primaryPropertyLoanToValue, 
    MAX(logPrimaryPropertyAvm) AS logPrimaryPropertyAvm, 
    MAX(hasTrust) AS hasTrust, 
    MAX(politicalDemocrat) AS politicalDemocrat, 
    MAX(politicalRepublican) AS politicalRepublican, 
    MAX(politicalIndependent) AS politicalIndependent, 
    MAX(femaleCount) AS femaleCount, 
    MAX(maleCount) AS maleCount, 
    MAX(minHouseholdAge) AS minHouseholdAge, 
    MAX(avgHouseholdAge) AS avgHouseholdAge, 
    MAX(maxHouseholdAge) AS maxHouseholdAge, 
    MAX(metroNameRank) AS metroNameRank, 
    MAX(propertyCount) AS propertyCount, 
    MAX(isSFROwner) AS isSFROwner, 
    MAX(isCondoOwner) AS isCondoOwner, 
    MAX(isRuralResidenceOwner) AS isRuralResidenceOwner, 
    MAX(logMaxPropertyAvm) AS logMaxPropertyAvm, 
    MAX(logSumPropertyAvm) AS logSumPropertyAvm, 
    MAX(maxPropertySqft) AS maxPropertySqft, 
    MAX(maxBedrooms) AS maxBedrooms, 
    MAX(primaryPropertySqft) AS primaryPropertySqft, 
    MAX(primaryPropertyYearsOwned) AS primaryPropertyYearsOwned, 
    MAX(minPropertyYearsOwned) AS minPropertyYearsOwned, 
    MAX(maxPropertyYearsOwned) AS maxPropertyYearsOwned, 
    MAX(logMaxDonationAmount) AS logMaxDonationAmount, 
    MAX(logSumDonationAmount) AS logSumDonationAmount, 
    MAX(isArtsCause) AS isArtsCause, 
    MAX(isEducationCause) AS isEducationCause, 
    MAX(isEnvironmentalCause) AS isEnvironmentalCause, 
    MAX(isAnimalCause) AS isAnimalCause, 
    MAX(isHealthCause) AS isHealthCause, 
    MAX(isHumanServicesCause) AS isHumanServicesCause, 
    MAX(isInternationalCause) AS isInternationalCause, 
    MAX(isSocialBenefitCause) AS isSocialBenefitCause, 
    MAX(isReligiousCause) AS isReligiousCause, 
    MAX(logsumArtsCause) AS logsumArtsCause, 
    MAX(logsumEducationCause) AS logsumEducationCause, 
    MAX(logsumEnvironmentalCause) AS logsumEnvironmentalCause, 
    MAX(logsumAnimalCause) AS logsumAnimalCause, 
    MAX(logsumHealthCause) AS logsumHealthCause, 
    MAX(logsumHumanServicesCause) AS logsumHumanServicesCause, 
    MAX(logsumInternationalCause) AS logsumInternationalCause, 
    MAX(logsumSocialBenefitCause) AS logsumSocialBenefitCause, 
    MAX(logsumReligiousCause) AS logsumReligiousCause, 
    MAX(is990Donation) AS is990Donation, 
    MAX(isCoopDonation) AS isCoopDonation, 
    MAX(isFECContribution) AS isFECContribution, 
    MAX(isStateContribution) AS isStateContribution, 
    MAX(logsum990Donation) AS logsum990Donation, 
    MAX(logsumCOOPDonation) AS logsumCOOPDonation, 
    MAX(logsumFECDonation) AS logsumFECDonation, 
    MAX(logsumStateContribution) AS logsumStateContribution, 
    MAX(logmax990Donation) AS logmax990Donation, 
    MAX(logmaxCOOPDonation) AS logmaxCOOPDonation, 
    MAX(logmaxFECDonation) AS logmaxFECDonation,
    MAX(logmaxStateContribution) AS logmaxStateContribution, 
    MAX(countNumCharities) AS countNumCharities, 
    MAX(logMaxDonationAmount_1year) AS logMaxDonationAmount_1year, 
    MAX(logSumDonationAmount_1year) AS logSumDonationAmount_1year, 
    MAX(logsumCOOPDonation_1year) AS logsumCOOPDonation_1year, 
    MAX(logsumFECDonation_1year) AS logsumFECDonation_1year, 
    MAX(logsumStateContribution_1year) AS logsumStateContribution_1year, 
    MAX(countNumCharities_1year) AS countNumCharities_1year, 
    MAX(logMaxDonationAmount_3year) AS logMaxDonationAmount_3year, 
    MAX(logSumDonationAmount_3year) AS logSumDonationAmount_3year, 
    MAX(logsum990Donation_3year) AS logsum990Donation_3year, 
    MAX(logsumCOOPDonation_3year) AS logsumCOOPDonation_3year, 
    MAX(logsumFECDonation_3year) AS logsumFECDonation_3year, 
    MAX(logsumStateContribution_3year) AS logsumStateContribution_3year, 
    MAX(countNumCharities_3year) AS countNumCharities_3year, 
    MAX(logMaxDonationAmount_5year) AS logMaxDonationAmount_5year, 
    MAX(logSumDonationAmount_5year) AS logSumDonationAmount_5year, 
    MAX(logsum990Donation_5year) AS logsum990Donation_5year, 
    MAX(logsumCOOPDonation_5year) AS logsumCOOPDonation_5year, 
    MAX(logsumFECDonation_5year) AS logsumFECDonation_5year, 
    MAX(logsumStateContribution_5year) AS logsumStateContribution_5year, 
    MAX(countNumCharities_5year) AS countNumCharities_5year, 
    MAX(lux_athletic) AS lux_athletic, 
    MAX(lux_flight) AS lux_flight, 
    MAX(lux_goods) AS lux_goods, 
    MAX(lux_travel) AS lux_travel, 
    MAX(num_vehicles) AS num_vehicles, 
    MAX(num_luxury_vehicles) AS num_luxury_vehicles, 
    MAX(num_ultra_luxury_vehicles) AS num_ultra_luxury_vehicles, 
    MAX(num_cars) AS num_cars, 
    MAX(num_trucks) AS num_trucks, 
    MAX(num_suvs) AS num_suvs, 
    MAX(num_vans) AS num_vans, 
    MAX(metroRank) AS metroRank, 
    MAX(stateRank) AS stateRank,
    --     c.*, 
    --     wf.* EXCEPT(id) 
FROM customer_features c 
LEFT JOIN wf ON c.windfall_id = wf.id
WHERE c.candidate_id IS NOT NULL
GROUP BY 1
"""
```

```python
import pandas as pd
from pandas import DataFrame, Series
import logging
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import (
    RandomForestClassifier,
    AdaBoostClassifier,
    GradientBoostingClassifier,
    ExtraTreesClassifier,
)


class ModelConfig:

    """
    Class to read, validate, fix and load model_config.yaml file into an object.

    Parameter:
        config: a dictionary that contains the raw information in model_config.yaml

    Variables:
        feature_lists: a dictionary of lists that includes feature lists for different types of treatments
        feature_maps: a feature map that is mainly used by zoolander to map features to their business names
        features: list of all features
        categorical_features: list of all categorical features
        numeric_features: list of all numeric features
        model: a dictionary that contains all the model runtime parameters in model_config.model key
        resampling: a dictionary that contains all the resampling parameters in model_config.resampling key
        feature_selection: a dictionary that contains all the feature_selection parameters in model_config.feature_selection key
        param_grid: a dictionary that contains all the param_grid parameters in model_config.param_grid key

    Methods:
        set_feature_lists(): processes the features information and constructs the feature lists.
            if a value is not provided, this method will impute the value with some default options
            - default imputer: constant
            - default encoder: ordinal_encoding
            - default scaler: no_scaling

        set_param_grid(): processes the param_grid information and returns the a list of models with their
            param_grid for hyper-parameter tuning

        set_resampling_params(): processes the resampling information and return the a dict of resampling options.
            if not provided, this function will set the default as None and resampling won't be performed.

        set_feature_selection_params(): processes the feature_selection information and return the a dict of
            feature_selection options. if not provided, this function will set the default as None and
            feature_selection won't be performed.

        set_model_params(): processes the model runtime parameters and return the a dict of
            model_params options. if not provided, this function will set some default values.

    """

    def __init__(self, config: dict):

        # feature params
        self.feature_lists = self.set_feature_lists(config)
        self.feature_maps = config["features"]
        self.features = list(config["features"].keys())
        self.categorical_features = list([key for key,val in config["features"].items() if val["type"] == "categorical"])
        self.numeric_features = list([key for key,val in config["features"].items() if val["type"] == "numeric"])
        # runtime model params
        self.model = self.set_model_params(config)
        # re-sampling params
        self.resampling = self.set_resampling_params(config)
        # feature_selection params
        self.feature_selection = self.set_feature_selection_params(config) # config["feature_selection"]
        # param grid
        self.param_grid = self.set_param_grid(config)
        # rfm features
        self.rfm = self.set_rfm_features(config)


    @staticmethod
    def set_feature_lists(config):

        assert "features" in config.keys(), "model_config.yaml must have features configurations!"
        assert config["features"], "The features key in model_config.yaml can not be null"

        feature_lists = {
            "numeric": {
                "imputation": {
                    "mean": [],
                    "median": [],
                    "most_frequent": [],
                    "constant": [],
                    "constant_zero": [],
                    "constant_one": []
                },
                "scaling": {
                    "min_max": [],
                    "standard": [],
                    "no_scaling" : []
                }
            },
            "categorical": {
                "imputation": {
                    "most_frequent": [],
                    "constant": [],
                    "constant_zero": [],
                    "constant_one": []
                },
                "encoding": {
                    "ordinal_encoding": [],
                    "label_encoding": [],
                    "onehot_encoding": []
                }
            }
        }

        for key, val in config["features"].items():

            if val["type"] == "numeric":

                if "imputation" in val:
                    if val["imputation"] in list(feature_lists["numeric"]["imputation"].keys()):
                        feature_lists["numeric"]["imputation"][val["imputation"]].append(key)
                    else:
                        feature_lists["numeric"]["imputation"]["constant"].append(key)
                else:
                    feature_lists["numeric"]["imputation"]["constant"].append(key)

                if "scaling" in val:
                    if val["scaling"] in list(feature_lists["numeric"]["scaling"].keys()):
                        feature_lists["numeric"]["scaling"][val["scaling"]].append(key)
                    else:
                        feature_lists["numeric"]["scaling"]["no_scaling"].append(key)
                else:
                    feature_lists["numeric"]["scaling"]["no_scaling"].append(key)

            elif val["type"] == "categorical":

                if "imputation" in val:
                    if val["imputation"] in list(feature_lists["categorical"]["imputation"].keys()):
                        feature_lists["categorical"]["imputation"][val["imputation"]].append(key)
                    else:
                        feature_lists["categorical"]["imputation"]["constant"].append(key)
                else:
                    feature_lists["categorical"]["imputation"]["constant"].append(key)

                if "encoding" in val:
                    if val["encoding"] in list(feature_lists["categorical"]["encoding"].keys()):
                        feature_lists["categorical"]["encoding"][val["encoding"]].append(key)
                    else:
                        feature_lists["categorical"]["encoding"]["ordinal_encoding"].append(key)
                else:
                    feature_lists["categorical"]["encoding"]["ordinal_encoding"].append(key)

            else:
                raise ValueError(f"feature type for column: {key} must be either numeric or categorical")

        return feature_lists

    @staticmethod
    def set_param_grid(config):

        assert "param_grid" in config.keys(), "model_config.yaml must have param_grid configurations!"
        assert config["param_grid"], "The param_grid key in model_config.yaml can not be null"
        rand_state = 42
        if "random_state" in config["model"].keys() and type(config["model"]["random_state"]) == int:
            rand_state = config["model"]["random_state"]

        model_param_list = []
        for model in config["param_grid"].keys():
            param_grid = {}

            # Set the model type 
            if model == "RandomForestClassifier":
                param_grid["clf__estimator"] = [RandomForestClassifier(random_state=rand_state)]
            elif model == "GradientBoostingClassifier":
                param_grid["clf__estimator"] = [GradientBoostingClassifier(random_state=rand_state)]
            elif model == "LogisticRegression":
                param_grid["clf__estimator"] = [LogisticRegression()]
            elif model == "DecisionTreeClassifier":
                param_grid["clf__estimator"] = [DecisionTreeClassifier(random_state=rand_state)]
            elif model == "KNeighborsClassifier":
                param_grid["clf__estimator"] = [KNeighborsClassifier()]
            elif model == "LinearDiscriminantAnalysis":
                param_grid["clf__estimator"] = [LinearDiscriminantAnalysis()]
            elif model == "GaussianNB":
                param_grid["clf__estimator"] = [GaussianNB()]
            elif model == "SVC":
                param_grid["clf__estimator"] = [SVC()]
            elif model == "AdaBoostClassifier":
                param_grid["clf__estimator"] = [AdaBoostClassifier(random_state=rand_state)]
            elif model == "ExtraTreesClassifier":
                param_grid["clf__estimator"] = [ExtraTreesClassifier(random_state=rand_state)]
            else:
                raise ValueError("The pipeline only supports the following values for model keys: "
                                 "[RandomForestClassifier, GradientBoostingClassifier, LogisticRegression, "
                                 "DecisionTreeClassifier, KNeighborsClassifier, LinearDiscriminantAnalysis,"
                                 "GaussianNB, SVC, AdaBoostClassifier, ExtraTreesClassifier]")
                
            # Add all hyper parmas
            for param in config["param_grid"][model]:
                param_grid[f"clf__estimator__{param}"] = config["param_grid"][model][param]
            
            # Append the model dict to the list
            model_param_list.append(param_grid)
            
        return model_param_list

    @staticmethod
    def set_resampling_params(config):

        if "resampling" not in config.keys() \
                or config["resampling"] is None:
            logging.info('resampling is not configured and will not be performed!')
            resampling_params = {"required": False}
            return resampling_params

        if ("required" not in config["resampling"].keys()) \
                or ("required" in config["resampling"].keys()
                    and config["resampling"]["required"] == False):
            resampling_params = {"required": False}
            return resampling_params

        resampling_params = config["resampling"]
        return resampling_params

    @staticmethod
    def set_feature_selection_params(config):

        feature_selection_params = dict()
        feature_selection_params["required"] = True
        feature_selection_params["model"] = ExtraTreesClassifier(n_estimators=200, random_state=42)
        feature_selection_params["threshold"] = None
        n_estimators = 200

        if "feature_selection" not in config.keys() \
                or config["feature_selection"] is None:
            logging.info('feature_selection is not configured and will be set to default!')
            return feature_selection_params

        # Set required param
        if "required" in config["feature_selection"].keys() \
                and type(config["feature_selection"]["required"]) == bool:
            feature_selection_params["required"] = config["feature_selection"]["required"]
        else:
            logging.info('feature_selection.required is not configured and will be set to default: True')

        # Set n_estimator param
        if "n_estimators" in config["feature_selection"].keys() \
                and type(config["feature_selection"]["n_estimators"]) in [float, int]:
            n_estimators = config["feature_selection"]["n_estimators"]
        else:
            logging.info('feature_selection.n_estimators is not configured and will be set to default: 200')

        # Set random_state
        if "random_state" in config["feature_selection"].keys() \
                and type(config["feature_selection"]["random_state"]) == int:
            random_state = config["feature_selection"]["random_state"]
        else:
            logging.info('feature_selection.random_state is not configured and will be set to default: 42')


        # Set threshold param
        if "threshold" not in config["feature_selection"].keys():
            logging.info('feature_selection.threshold is not configured properly and will be set to default value: None')
        elif type(config["feature_selection"]["threshold"]) == str \
            and config["feature_selection"]["threshold"] not in ['median', 'mean', 'None']:
            logging.info('feature_selection.threshold is not configured properly and will be set to default value: None')
        elif config["feature_selection"]["threshold"] in [None, 'None']:
            logging.info('feature_selection.threshold is set to default value: None')
        else:
            feature_selection_params["threshold"] = config.feature_selection["threshold"]

        # Set estimator/model param
        if "model" not in config["feature_selection"].keys():
            logging.info('feature_selection.model is not configured and will be set to default value: ExtraTreesClassifier')
        elif config["feature_selection"]["model"] == 'RandomForestClassifier':
            feature_selection_params["model"] = RandomForestClassifier(n_estimators=n_estimators,
                                                                       random_state=random_state)
        else:
            logging.info('feature_selection.threshold is set to default value: ExtraTreesClassifier')
            feature_selection_params["model"] = ExtraTreesClassifier(n_estimators=n_estimators,
                                                                     random_state=random_state)

        return feature_selection_params

    @staticmethod
    def set_model_params(config):

        model_params = dict()
        model_params["version"] = "0.0.1"
        model_params["stratified"] = True
        model_params["random_grid_search"] = True
        model_params["random_iter"] = 100
        model_params["num_cv"] = 5
        model_params["model_scoring"] = "average_precision"
        model_params["random_state"] = 42
        model_params["confusion_threshold"] = 0.5

        if "model" not in config.keys() \
                or config["model"] is None:
            logging.info('config.model is not configured and will be set to default!')
            return model_params

        if "version" in config["model"].keys():
            model_params["version"] = config["model"]["version"]
        if "stratified" in config["model"].keys() and type(config["model"]["stratified"]) == bool:
            model_params["stratified"] = config["model"]["stratified"]
        if "random_grid_search" in config["model"].keys() and \
                type(config["model"]["random_grid_search"]) == bool:
            model_params["random_grid_search"] = config["model"]["random_grid_search"]
        if "random_iter" in config["model"].keys() and type(config["model"]["random_iter"]) == int:
            model_params["random_iter"] = config["model"]["random_iter"]
        if "num_cv" in config["model"].keys() and type(config["model"]["num_cv"]) == int:
            model_params["num_cv"] = config["model"]["num_cv"]
        if "model_scoring" in config["model"].keys() and type(config["model"]["model_scoring"]) == str:
            model_params["model_scoring"] = config["model"]["model_scoring"]
        if "random_state" in config["model"].keys() and type(config["model"]["random_state"]) == int:
            model_params["random_state"] = config["model"]["random_state"]
        if "confusion_threshold" in config["model"].keys() and \
                type(config["model"]["confusion_threshold"]) == float:
            model_params["confusion_threshold"] = config["model"]["confusion_threshold"]

        return model_params

    @staticmethod
    def set_rfm_features(config):

        rfm = dict()
        rfm["r"] = "recent_donation_months_before_target"
        rfm["f"] = "five_yr_prior_ct_rfm"
        rfm["m"] = "five_yr_prior_amt_rfm"
        rfm["y_amount"] = "target_period_amt"

        if "rfm" not in config.keys() \
                or config["rfm"] is None:
            logging.info('config.rfm is not configured and will be set to default!')
            return rfm
        if "r" in config["rfm"].keys():
            rfm["r"] = config["rfm"]["r"]
        if "f" in config["rfm"].keys():
            rfm["f"] = config["rfm"]["f"]
        if "m" in config["rfm"].keys():
            rfm["m"] = config["rfm"]["m"]
        if "y_amount" in config["rfm"].keys():
            rfm["y_amount"] = config["rfm"]["y_amount"]
        return rfm
```
